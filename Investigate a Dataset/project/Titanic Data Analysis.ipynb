# Background Information

The sinking of the **RMS Titanic** occurred on the night of 14 April through to the morning of 15 April 1912 in the North Atlantic Ocean, four days into the ship's maiden voyage from Southampton to New York City. The largest passenger liner in service at the time, Titanic had an estimated 2,224 people on board when she struck an iceberg at around 23:40 (ship's time) on Sunday, 14 April 1912. Her sinking two hours and forty minutes later at 02:20 (05:18 GMT) on Monday, 15 April resulted in the deaths of more than 1,500 people, which made it one of the deadliest peacetime maritime disasters in history. ([Wikipedia](https://en.wikipedia.org/wiki/Sinking_of_the_RMS_Titanic))

One of the reasons that the shipwreck led to such loss of life was that there were not enough lifeboats for the passengers and crew. Although there was some element of luck involved in surviving the sinking, some groups of people were more likely to survive than others, such as women, children, and the upper-class. ([Kaggle](https://www.kaggle.com/c/titanic))

# Data Wrangling

### Drop unused data

````python
# drop columns `PassengerId`, `Name`, `Ticket`, `Fare`, and `Cabin` 
titanic_df = titanic_df.drop(['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin'], axis=1)
````

### Remove missing data

````python
# remove missing data from 'Age' and 'Embarked' that contain some missing data
titanic_df.dropna(axis=0, inplace= True)
````

### Store the `Survived` feature

````python
# Store the 'Survived' feature in a new variable and remove it from the dataset 
outcomes = titanic_df['Survived']
data = titanic_df.drop('Survived', axis = 1)
````

# Data Analysis

## 1. Passenger Class
Is buying a 1st class ticket can increase the chances of survival?

**Answer:**

1st (Pclass = 1) class passengers have the biggest survival rate compare the others. So, it could increase the chances of survival.

````python
# Exploring the passenger class data
df_Pclass = pd.crosstab(titanic_df.Pclass, titanic_df.Survived)
df_Pclass['Total'] = df_Pclass[0] + df_Pclass[1]
df_Pclass
````

````python
# probability of survival for each class
class1 = (float(df_Pclass[1][1])/(df_Pclass[0][1] + df_Pclass[1][1]))*100
class2 = (float(df_Pclass[1][2])/(df_Pclass[0][2] + df_Pclass[1][2]))*100
class3 = (float(df_Pclass[1][3])/(df_Pclass[0][3] + df_Pclass[1][3]))*100
print("Probability of survival for each class:")
print("Class 1: {0:.2f}%".format(class1))
print("Class 2: {0:.2f}%".format(class2))
print("Class 3: {0:.2f}%".format(class3))

# visualize in pie chart for comparison all of classes
percentage_class1 = class1/(class1 + class2 + class3)
percentage_class2 = class2/(class1 + class2 + class3)
percentage_class3 = class3/(class1 + class2 + class3)

# Data to plot
labels = 'Class 1', 'Class 2', 'Class 3'
sizes = [percentage_class1, percentage_class2, percentage_class3]
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Passenger Class Survival Percentage', bbox={'facecolor':'0.8', 'pad':5}, y=1.08)
plt.axis('equal')

print("\nComparison all of the classes that has the biggest chance to survive:")
plt.show()
````

````python
vs.survival_stats(data, outcomes, 'Pclass')
````

## 3. Sex

Who has a higher chance of survival between male and female?

**Answer:**
Female passengers have the biggest survival rate than male passengers.

````python
# Exploring the passenger class data
df_Sex = pd.crosstab(titanic_df.Sex, titanic_df.Survived)
df_Sex['Total'] = df_Sex[0] + df_Sex[1]
df_Sex
````

````python
# probability of survival for males and females
males = (float(df_Sex[1][1])/(df_Sex[0][1] + df_Sex[1][1]))*100
females = (float(df_Sex[1][0])/(df_Sex[0][0] + df_Sex[1][0]))*100
print("Probability of survival for males and females:")
print("Male: {0:.2f}%".format(males))
print("Female: {0:.2f}%".format(females))

# visualize in pie chart for comparison 
percentage_males = males/(males + females)
percentage_females = females/(males + females)

# Data to plot
labels = 'Males', 'Females'
sizes = [percentage_males, percentage_females]
colors = ['gold', 'yellowgreen']
explode = (0.1, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Sex Survival Percentage', bbox={'facecolor':'0.8', 'pad':5})
plt.axis('equal')

print("\nComparison for males and females that has the biggest chance to survive:")
plt.show()
````

````python
vs.survival_stats(data, outcomes, 'Sex')
````

## 6. Embarked

What are the chances of survival of passengers departing from Cherbourg?

**Answer:**
Passengers embarked at Cherbourg have the bigger survival rate that others.

````python
# Exploring the passenger class data
df_Embarked = pd.crosstab(titanic_df.Embarked, titanic_df.Survived)
df_Embarked['Total'] = df_Embarked[0] + df_Embarked[1]
df_Embarked
````

````python
# probability of survival
cherbourg = (float(df_Embarked[1][0])/(df_Embarked[0][0] + df_Embarked[1][0]))*100
queenstown = (float(df_Embarked[1][1])/(df_Embarked[0][1] + df_Embarked[1][1]))*100
southhampton = (float(df_Embarked[1][2])/(df_Embarked[0][2] + df_Embarked[1][2]))*100
print("Probability of survival:")
print("Class 1: {0:.2f}%".format(cherbourg))
print("Class 2: {0:.2f}%".format(queenstown))
print("Class 3: {0:.2f}%".format(southhampton))

# visualize in pie chart for comparison all of classes
percentage_cherbourg = cherbourg/(cherbourg + queenstown + southhampton)
percentage_queenstown = queenstown/(cherbourg + queenstown + southhampton)
percentage_southhampton = southhampton/(cherbourg + queenstown + southhampton)

# Data to plot
labels = 'Cherbourg', 'Queenstown', 'Southhampton'
sizes = [percentage_cherbourg, percentage_queenstown, percentage_southhampton]
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Ports Embarked Survival Percentage', bbox={'facecolor':'0.8', 'pad':5}, y=1.08)
plt.axis('equal')

print("\nComparison all of the ports embarked that has the biggest chance to survive:")
plt.show()
````
